```python
import requests
import json
import csv
import pandas as pd
import numpy as np
```

```python
data0 = requests.get("http://api.alquran.cloud/v1/juz/1/quran-uthmani")
data1 = requests.get("http://api.alquran.cloud/v1/juz/2/quran-uthmani")
data2 = requests.get("http://api.alquran.cloud/v1/juz/3/quran-uthmani")
```

```python
result0 = json.loads(data0.text)
result1 = json.loads(data1.text)
result2 = json.loads(data2.text)
result0
```

```python
surahs = []
for i in result0['data']['ayahs']:
    surahs.append(i if i["surah"]["name"] == 'سُورَةُ البَقَرَةِ' else np.nan)

for i in result1['data']['ayahs']:
    surahs.append(i if i["surah"]["name"] == 'سُورَةُ البَقَرَةِ' else np.nan)
    
for i in result2['data']['ayahs']:
    surahs.append(i if i["surah"]["name"] == 'سُورَةُ البَقَرَةِ' else np.nan)

df_s = pd.DataFrame(surahs)
data_surahs = df_s.dropna()
for i in data_surahs[0]:
    i['number'] -= 7
data_surahs.to_csv('Al-Baqara.csv', header=False, index=False)
data_surahs
```

```python
list_number_of_surahs = []
list_text_of_surahs = []
for i in data_surahs[0]:
    list_number_of_surahs.append(i['number']-7)
    list_text_of_surahs.append(i["text"])
dict_surahs = {'text':list_text_of_surahs, 'number':list_number_of_surahs}
df2 = pd.DataFrame(dict_surahs)
df2
```

```python
df2.to_csv('Baqara.csv', header=False, index=False)
```

```python
with open('Baqara.csv', 'r',encoding="utf8") as f_csv, open('Baqara.txt', 'w', encoding="utf-8") as f_txt:
    surah_list = f_csv.read().split('\n')
    str_s = ' '.join([str(elem) for elem in surah_list])
    result_surah = ' '.join([str(elem) for elem in str_s.split(",")])
    f_txt.write(result_surah)
```
