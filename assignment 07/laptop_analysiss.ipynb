```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("laptop_pricing_dataset_mod1.csv", header=None)
df.head()
```

```python
headers = ['index','Manufacturer','Category','Screen','GPU','OS','CPU_core','Screen_Size_cm',
           'CPU_frequency','RAM_GB','Storage_GB_SSD','Weight_kg','Price']
df.columns = headers
df.head()
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df.dtypes 
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_norm_loss = df["Weight_kg"].astype("float").mean(axis=0)
df["Weight_kg"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
for column in df[["Screen_Size_cm"]]:
    print(df[column].value_counts())
    print('The most repeated value: ', df[column].value_counts().idxmax())
    
df["Screen_Size_cm"].replace(np.nan, df[column].value_counts().idxmax(), inplace=True)
```

```python
df['Weight_kg'] = 2.20462*df["Weight_kg"]
df.rename(columns={'Weight_kg':'Weight_pound'}, inplace=True)
```

```python
df['Screen_Size_cm'] = 0.393701*df["Screen_Size_cm"]
df.rename(columns={'Screen_Size_cm':'Screen_Size_inch'}, inplace=True)
```

```python
for column in df[["CPU_frequency"]]:
    print('The most repeated value: ', df[column].value_counts().idxmax())
    
df["CPU_frequency"] = df[column].value_counts().idxmax() // df["CPU_frequency"]
```

```python
for column in df[["CPU_frequency"]]:
    print(df[column].value_counts())
```

```python
bins = np.linspace(min(df["Price"]), max(df["Price"]), 4)
bins
```

```python
group_names = ["Low","Medium","High"]
df['Price-binned'] = pd.cut(df['Price'], bins, labels=group_names,  include_lowest=True)
df[['Price','Price-binned']].head()
```

```python
for column in df[["Price-binned"]]:
    print(df[column].value_counts())
```

```python
Full_HD = []
IPS_Panel = []
for column in df["Screen"]:
    if column == 'IPS Panel':
        IPS_Panel.append(1)
    else:
        IPS_Panel.append(0)
    if column == 'Full HD':
        Full_HD.append(1)
    else:
        Full_HD.append(0)
df['Screen-IPS_Panel'] = IPS_Panel
df['Screen-Full_HD'] = Full_HD
```
