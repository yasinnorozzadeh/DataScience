```python
import pandas as pd
```

```python
df = pd.read_csv('forbes_2640_billionaires.csv')

df.head()
```

```python
df.describe(include='all') 
```

```python
industry = df.loc[:]["industry"]
industry
```

```python
c_industry = industry.value_counts()
"Most people's job fields : {}".format(c_industry.index[0])
```

```python
x = 0
for i in range(10):
    x += float(df.iloc[i,3])
```

```python
print('Average net worth of all people: {}'.format(df['net_worth'].describe().mean()))
print('Average net worth of top 10 people: {}'.format(x/10))
```

```python
'The difference between the average total wealth and the average wealth of the top 10 people: {}'.format(df['net_worth'].describe().mean() - x/10)
```
